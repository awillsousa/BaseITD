```python
import os

INDICADOR_DIV_PG = "page-break-before:always; page-break-after:always"
arquivo_acordao = ARQUIVO.replace('.html','')
#DIRETORIO_ACORDAOS = "./separados/html/{0}".format(arquivo_acordao)
DIRETORIO_ACORDAOS = os.path.join(PATH_BASE_DATA, "separados",arquivo_acordao,"html")

# Ementa e acordao
##eme_acor_html = parse_html("{0}/{1}_Ementa_e_Acordao.html".format(DIRETORIO_ACORDAOS,arquivo_acordao))
##relatorio_html = parse_html("{0}/{1}_Relatorio.html".format(DIRETORIO_ACORDAOS,arquivo_acordao))
eme_acor_html = None
relatorio_html = None
votos_html = {}

for r, d, f in os.walk(DIRETORIO_ACORDAOS):
    for file in f:        
        if '.html' in file:
            if '_Voto_' in file:
                votos_html[file.replace('.html','')] = parse_html(os.path.join(r, file))
            elif '_Ementa_e_Acordao_' in file:
                eme_acor_html = parse_html(os.path.join(r, file))
            elif '_Relatorio_' in file:
                relatorio_html = parse_html(os.path.join(r, file))
            #files.append(os.path.join(r, file))

#votos_html = {'6_Voto_MIN_EDSON_FACHIN': parse_html("separados/6/6_Voto_MIN_EDSON_FACHIN.html"),
#              '6_Voto_MIN_MARCO_AURELIO': parse_html("separados/6/6_Voto_MIN_MARCO_AURELIO.html")}

acordao = {}

acordao['acordao_ementa_html'] = eme_acor_html
acordao['relatorio_html'] = relatorio_html
acordao['partes_html'] = []
acordao['votos_html'] = votos_html

acordao['descricao'] = ''
acordao['acordao'] = []
acordao['ementa'] = []
acordao['relatorio'] = []
acordao['votos'] = {f:[] for f in votos_html.keys()}
acordao['partes'] = {}

# Processa as paginas relativas ao acordao e ementa
def processa_acordao_ementa(acordao):
    # cada div corresponde a uma página
    divs = []    
    divs.extend(acordao['acordao_ementa_html'].find_all("div",  style=INDICADOR_DIV_PG))
    
    # Itera as divs e seus paragrafos para separar as partes, ementa e texto do acordao
    flag_pag1 = True
    parte_atual = "PARTES"
    for div in divs:
        for i,parag in enumerate(div.find_all("p")): 
                print("Parte atual: [ {2} ] Tag <p> sendo processada: {0}-{1}".format(i,parag.text,parte_atual))
                if i == 0 or i == 1:
                    parag.decompose()
                elif i==2 and flag_pag1:                    
                    flag_pag1 = False
                    parag.decompose()
                elif any([r.match(parag.text.replace('\n',' ')) for r in padroes_excluir]) or any([r.match(parag.text) for r in padroes_excluir]):
                    parag.decompose()
                else:
                    #print("Parte atual: [ {2} ] Tag <p> sendo processada: {0}-{1}".format(i,parag.text,parte_atual))
                    if "A C Ó R D Ã O" in parag.text or "ACÓRDÃO" in parag.text:
                        if not any(pOK in parag.text for pOK in PARTES_OK):
                            # Nesta parte, depois será bom verificar se sempre acontece do indicador do acórdão
                            # "A C Ó R D Ã O" vir dentro da tag <p> anterior                    
                            parag.string = parag.text.replace("A C Ó R D Ã O",'') 
                            parag.string = parag.text.replace("ACÓRDÃO",'')
                            parte_atual = "ACORDAO"                        
                            acordao['ementa'].append(parag)
                        elif parte_atual == "PARTES":
                            acordao['partes_html'].append(parag) 
                            
                    elif "EMENTA" in parag.text.replace(' ','').upper():
                        parte_atual = "EMENTA"                        
                        acordao['ementa'].append(parag)
                    else:
                        if parag:  # Vai saber se não tem None na jogada                                                        
                            if parte_atual == "PARTES":                                                                
                                print("Encontrado {}".format(parag.text))
                                acordao['partes_html'].append(parag)                                                                 
                            elif parte_atual == "EMENTA":                                
                                acordao['ementa'].append(parag)
                            elif parte_atual == "ACORDAO":                                
                                acordao['acordao'].append(parag)                
    
    acordao['ementa'] = ' '.join([s.text.replace('\n', ' ') for s in acordao['ementa']])
    acordao['acordao'] = ' '.join([s.text.replace('\n', ' ') for s in acordao['acordao']])
    
    return (acordao)
                    
                    
                      
                    #parag.string = parag.text.replace('\n',' ')

# Processa as paginas relativas ao relatório
def processa_relatorio(acordao):
    # cada div corresponde a uma página
    divs = acordao['relatorio_html'].find_all("div",  style=INDICADOR_DIV_PG)
    
    # Itera as divs e seus paragrafos para extrair apenas o texto do relatório    
    
    flag_pag1 = True
    parte_atual = "PARTES"
    for div in divs:        
        for i,parag in enumerate(div.find_all("p")): 
                if i == 0 or i == 1:
                    parag.decompose()
                elif i==2 and flag_pag1:                    
                    flag_pag1 = False
                    parag.decompose()
                
                if "R E L A T Ó R I O" in parag.text or "RELATÓRIO" in parag.text:
                    parte_atual = "RELATORIO" 
                elif any([r.match(parag.text.replace('\n',' ')) for r in padroes_excluir]):
                    parag.decompose()
                elif parte_atual == "RELATORIO":
                    if parag:
                        acordao['relatorio'].append(parag)                    
    
    acordao['relatorio'] = ' '.join([s.text.replace('\n', ' ') for s in acordao['relatorio']])
    
    return (acordao)                             
                            
    
    
# Processa todos os votos do acordão
def processa_votos(acordao):
    for voto, texto_voto in acordao['votos_html'].items():
        # cada div corresponde a uma página         
        divs = texto_voto.find_all("div",  style=INDICADOR_DIV_PG)

        # Itera as divs e seus paragrafos para extrair apenas o texto do relatório        
        flag_pag1 = True
        for div in divs:            
            for i,parag in enumerate(div.find_all("p")): 
                if i == 0 or i == 1:
                    parag.decompose()
                elif i==2 and flag_pag1:                    
                    flag_pag1 = False
                    parag.decompose()
                
                if "V O T O" in parag.text:
                    parte_atual = "VOTO" 
                elif any([r.match(parag.text.replace('\n',' ')) for r in padroes_excluir]):
                    parag.decompose()
                else: # parte_atual == "VOTO"
                    if parag:                        
                        acordao['votos'][voto].append(parag)                    
        
        acordao['votos'][voto] = ' '.join([s.text.replace('\n', ' ') for s in acordao['votos'][voto]])
        
    return (acordao)


PADROES_SIGLA_EXCLUIR = ['PROCURADOR-GERAL', 'ADVOGADO-GERAL']
def processa_partes(acordao):
    partes_display = "".join([p.text for p in acordao['partes_html']])
    linhas_parte = partes_display.split('\n')
    linhas_excluir = []
    linha_marcada = 0
    residual = ""
    tem_residual = False
    iter_partes = linhas_parte[:]
    for i,linha in enumerate(iter_partes):
        if ":" not in linha:
            print("Parte da linha anterior " + linha)
            if any(pOK in linha for pOK in PARTES_OK) or tem_residual:
                tem_residual = True
                residual += linha
                residual += " "                
            else:                                       
                linhas_parte[linha_marcada] += linha
                
            linhas_excluir.append(i)    
        else:            
            linha_marcada = i
            
            if tem_residual:
                print("Linha marcada "+linha)
                print("Residual "+residual)
                linhas_parte[linha_marcada] = residual + linha 
                tem_residual = False
                residual = ""
            
    iter_partes = None # Gambiaaaaarra!
    for l in linhas_excluir[::-1]:
        del linhas_parte[l]

    partes = {}    
    for linha in linhas_parte:
        l = linha.split(':')        
        partes[l[0].replace(' ','')] = l[1]
    
    #acordao_partes = partes
    acordao['partes'] = []
    for tipo, nome in partes.items():
        nome_base = ''
        sigla = ''
        if '-' in nome and not any(f in nome for f in PADROES_SIGLA_EXCLUIR):
            l_nome_base = nome.split('-')
            nome_base = '-'.join(l_nome_base[:-1])    
            sigla = l_nome_base[-1].replace(' ','')
        elif 'E OUTRO(A/S)' in nome:
            nome_base = nome
            nome_base.replace('E OUTRO(A/S)', '')
        else:
            nome_base = nome
        
        tipo_base = tipo.replace(' ','')
        acordao['partes'].append({'tipo':tipo_base.lstrip(' ').rstrip(' '), 
                                  'nome': nome_base.lstrip(' ').rstrip(' '), 
                                  'sigla': sigla.lstrip(' ').rstrip(' ')})
    
    
    return(acordao)
            
# Executa os processamentos das partes do acordao
acordao = processa_acordao_ementa(acordao)
print("--- EMENTA ---\n\n" +str(acordao['ementa'])+"\n\n")
print("--- ACORDAO ---\n\n"+str(acordao['acordao'])+"\n\n")

acordao = processa_partes(acordao)
print("--- PARTES ---\n\n" +str(acordao['partes'])+"\n\n")

acordao = processa_relatorio(acordao)
print("--- RELATORIO ---\n\n"+str(acordao['relatorio'])+"\n\n")

acordao = processa_votos(acordao)
print("--- VOTOS ---\n\n"+str(acordao['votos'])+"\n\n")

```

### Testes Simples de NLP


```python
from nltk.tokenize import sent_tokenize, word_tokenize, RegexpTokenizer
from nltk import download, pos_tag, FreqDist, data, corpus
from nltk.corpus import floresta
from nltk.util import ngrams

import string
import re

#download()  # na primeira execução necessitamos baixar alguns pacotes (rslp, all-book)

acordaoqlq_txt = acordao['ementa'] + acordao['acordao']+acordao['relatorio']

tokens = word_tokenize(acordaoqlq_txt)
stopwords = corpus.stopwords.words('portuguese')
   
# tonekiza sentencas (linhas) do texto
stok = data.load('tokenizers/punkt/portuguese.pickle')        
sentencas = stok.tokenize(acordaoqlq_txt)
tokenizer = RegexpTokenizer(r'\w+')
palavras_nosw = []
for s in sentencas:    
    t_s = tokenizer.tokenize(s.lower())
    palavras_nosw += [w for w in t_s if not w in stopwords and len(w) > 1]

texto_filtrado = " ".join(palavras_nosw)
tokens = tokenizer.tokenize(texto_filtrado)
list(tokens)
```

```python
# bigramas e trigramas
duplas = ngrams(tokens,2)
palavras_duplas = []
for b in list(duplas):
    palavras_duplas.append(' '.join(b))

trios = ngrams(tokens,3)
palavras_triplas = []
for t in list(trios):
    palavras_triplas.append(' '.join(t))

n=40

print("\n{0} - {2} primeiras (total = {1})".format("PALAVRAS", len(palavras_nosw), n))
print(palavras_nosw[:n])
print("\n{0} - {2} primeiras (total = {1})".format("PALAVRAS DUPLAS", len(palavras_duplas), n))
print(palavras_duplas[:n])
print("\n{0} - {2} primeiras (total = {1})".format("PALAVRAS TRIPLAS", len(palavras_triplas), n))
print(palavras_triplas[:n])
```

```python
# contagem de frequencia dos termos
termos = list(set(palavras_duplas)) + list(set(palavras_triplas))
fdist = {}

for m in termos:
    fdist[m] = len(re.findall(m, texto_filtrado))

freqs = []
for m in termos:
    #if fdist[m] > 0 and "voto" in m:
    if fdist[m] > 1:
        freqs.append('{}: {}'.format( m, fdist[m] ))

print("FREQUENCIA DE n-GRAMAS")        
sorted(freqs)
```
